```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from matplotlib import style
```

```python
x_data = np.array([0,1,2,3,1,1.5])
y_data = np.array([1,2,1.5,2,1,2])
```

```python
def model_function(x,alpha_0,alpha_1,alpha_2):
    res = alpha_0*(x-alpha_1)**2 + alpha_2
    return res
```

```python
optimal_para,covariance_mat = curve_fit(model_function,x_data,y_data)
print(f"optimal alpha_0 = {optimal_para[0]}")
print(f"optimal alpha_1 = {optimal_para[1]}")
print(f"optimal alpha_2 = {optimal_para[2]}")
alpha_0,alpha_1,alpha_2 = optimal_para
```

```python
style.use('seaborn')
plt.scatter(x_data,y_data,label = 'data point')
x_temp = np.linspace(min(x_data), max(x_data), 80)
y_model = model_function(x_temp,alpha_0,alpha_1,alpha_2)
plt.plot(x_temp,y_model,label = 'best fit line',color='red',linewidth=3)
plt.title('polynomial curve fitting')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.show()
```
