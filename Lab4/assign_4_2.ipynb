```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
x_values = np.random.rand(6)*3+1
y_values = 1.7+ 0.5*x_values
print(x_values)
print(y_values)
```

```python
alpha_0 = 0
alpha_1 = 0
l = 0.001
```

```python
n = len(x_values)
iterations = 3
```

```python
n = float(n)
style.use('seaborn')
for i in range(iterations):
    y_pred = alpha_0*x_values + alpha_1
    cost = sum((y_values-y_pred)**2)
    Dm = -(2/n) * (sum((y_values - y_pred) * x_values))
    Dc = -(2/n) * (sum(y_values - y_pred))
    alpha_0 = alpha_0 - l*Dm
    alpha_1 = alpha_1 - l*Dc
    y_pred = alpha_0*x_values + alpha_1
    print(f"cost = {cost}")
    print(f"alpha_0 = {alpha_0}")
    print(f"alpha_1 = {alpha_1}\n")
    plt.scatter(x_values,y_values,label = 'Data point',color ='green',s=80)
    plt.title(f'Iteration :- {i} cost = {cost}')
    plt.xlabel("x_values")
    plt.ylabel("y_values")
    plt.plot(x_values,y_pred,label = 'regression line',color='red',linewidth=3)
    plt.legend()
    plt.show()
    
```
